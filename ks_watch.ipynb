# Project definition/frame

**Business opportunities**

B1. Digital influence measurement both qualitative and quantitative made available to an influencer

B2. Introduction of an alternative qualification/quantification of a public digital influence: based on the assessment of the influencer self-designed (not openly) goals.

B3. Availability of a dictionnary of terms specific to the sub-culture crossing *nouchi, rap ivoire, ivorian humor*. Ivorian urban/artistic subculture currently has a significant impact among the french-speaking african youth and diaspora (potentially several hundreds milions of people)

B4. Automatic classification of jargon-filled comments


**Objectives**

O1. Clean the raw comments extracted from targeted facebook groups

O2. Achieve an EDA of the text comments  

O3. Perform a dynamic topic analysis of the comments and possibly confront with the evolution of the group hashtags in time (if those statistics are available to the group admins, and if they grant access to it).

O4. Make it possible to automatically classify the members comments (espicially those not self-hashtagged), based on comments topics generated.

**Assumptions**

A1. There is a high interconnection within the mediatic world which displays the facts and highlights of this subculture: important web-hosted or TV shows litterally relay down to everyday buzzes or gossips booming the mediatic sphere.

A2. There is a common slang-like jargon spanning the whole subculture.

A3. The ivorian urban/artistic subculture has the strongest impact in the african french-speaking youth, originating mainly from urban music and humoristic production.

**Update of the tactics**

The targeted groups are enlarged at least for the achievement of objectives O3 and O4. This is due to (1) a relative scarcity of text comments among the posts observed on the initial targets (initiated by the influencer staff), (2) the assumptions A1 and A2.

**Deliverables**

D1. A notebook of the EDA of text comments

D2. A dictionnary of a subculture jargon

D3. Sets of comments topics 

D4. A notebook of pipeline of comments processing
